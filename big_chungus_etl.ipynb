From here we can merge the tables with common attributes to simplify cleaning. First we'll join the tables that contain imdb's "tconst" index and a one to one relationship - that ends up being the title_basics_df and the title_ratings_df.

```python
#imdb_title_crew_df.info()
#imdb_title_ratings_df.info()
imdb_title_principals_df.info()
# imdb_title_basics_df.info()
```

Since title pincipals has the most entries (1,028,186) we'll use that as our foundation. Title basics and title crew have the same amount of entries so let's see which produces more from an inner join. 

```python
imdb_title_basics_ratings = imdb_title_basics_df.join(imdb_title_ratings_df, how='left')
imdb_title_basics_ratings.info()
```

```python
#Check for n/a values
imdb_title_basics_ratings.isna().sum()
```

Now we see that the imdb_crew_df has the same number of entries as our new df so let's merge those.

```python
imdb_basics_crew_ratings = imdb_title_basics_ratings.join(imdb_title_crew_df, how='outer')
imdb_basics_crew_ratings.info()
```

```python
#Let's check to make sure the index values are unique and don't have any duplicates
len(imdb_basics_crew_ratings.index.unique())
```

Great! Now it seems like we'll want to split up some of the values in the genres, writers, and directors columns:

```python
#genre is a list of upto 3 genres, splitting it to get the individual genres
imdb_basics_crew_ratings['genres'] = imdb_basics_crew_ratings['genres'].fillna('n/a')
imdb_basics_crew_ratings['genres'] = imdb_basics_crew_ratings['genres'].apply(lambda x: x.split(",") if x else x)
imdb_basics_crew_ratings['genres'].unique
```

```python
#making a set of all genres we have. set doesn't allow duplicate values

all_genres = set()
for genres in imdb_basics_crew_ratings['genres']:
    if genres:
        all_genres.update(genres)
```

```python
all_genres
```

```python
#adding cols with zeros for all the genres we have. Will modify genre to 1 if the film is of that genre.

for genre in all_genres:
    imdb_basics_crew_ratings[genre] = np.zeros(shape=imdb_basics_crew_ratings.shape[0])
                                
imdb_basics_crew_ratings.head()
```

```python
#setting the genre to be 1 if the film is of that genre

for index, row in imdb_basics_crew_ratings.iterrows():
    if row['genres']:
        for genre in row['genres']:
            imdb_basics_crew_ratings.loc[index, genre] = 1

imdb_basics_crew_ratings.head()
```

```python
movie_details_df_modified = imdb_basics_crew_ratings.drop(columns = 'genres')
```

```python

#checking the counts for all different genres

for col in movie_details_df_modified:
    print(f'Viewing values in col: {col}')
    print(f'Top 5 values:\n{movie_details_df_modified[col].value_counts()}')
```

```python
#making a list of all genres

cols = list(imdb_basics_crew_ratings.columns)
```

```python
genre_cols = cols[11:]
```

```python
#getting a dict with genre counts

genre_count = {}
for col in genre_cols:
    count = np.sum(imdb_basics_crew_ratings[col] == 1).sum() 
    genre_count[col] = count
```

```python
movie_details_df_modified.head()
```

Wonderful, now we have the genres split into usable data. Let's do something similar for writer and directors. 

```python
movie_details_df_modified['directors'] = movie_details_df_modified['directors'].fillna('n/a')
movie_details_df_modified['directors'] = movie_details_df_modified['directors'].apply(lambda x: x.split(",") if x else x)
movie_details_df_modified.head()
```

```python
# This function takes the lists of values in the director's category and creates a new dataframe of those values
f = lambda x: 'director_{}'.format(x + 1)

director_frame = pd.DataFrame(
    movie_details_df_modified['directors'].values.tolist(),
    movie_details_df_modified.index, dtype=object
).fillna('n/a').rename(columns=f)
```

```python
d1 = director_frame['director_1']
d2 = director_frame['director_2']
d3 = director_frame['director_3']
directors = pd.concat([d1, d2, d3], axis = 1)
directors.head()
```

```python
movie_details_df_modified = movie_details_df_modified.join(directors)
movie_details_df_modified.info()
```

```python
movie_details_df_modified = movie_details_df_modified.drop(columns = 'directors')
movie_details_df_modified.head()
```

Now let's do the same thing with writers:

```python
movie_details_df_modified['writers'] = movie_details_df_modified['writers'].fillna('n/a')
movie_details_df_modified['writers'] = movie_details_df_modified['writers'].apply(lambda x: x.split(",") if x else x)
movie_details_df_modified.head()
```

```python
# This function takes the lists of values in the director's category and creates a new dataframe of those values
f = lambda x: 'writer_{}'.format(x + 1)

writers_frame = pd.DataFrame(
    movie_details_df_modified['writers'].values.tolist(),
    movie_details_df_modified.index, dtype=object
).fillna('n/a').rename(columns=f)
```

```python
w1 = writers_frame['writer_1']
w2 = writers_frame['writer_2']
w3 = writers_frame['writer_3']
writers = pd.concat([w1, w2, w3], axis = 1)
writers.head()
```

```python
movie_details_df_modified = movie_details_df_modified.join(writers)
movie_details_df_modified.info()
```

```python
movie_details_df_modified = movie_details_df_modified.drop(columns = 'writers')
movie_details_df_modified.head()
```

This is what we want our frame to look like but now I'd like to swap the n/a's we put in for Nan

```python
big_chungus = movie_details_df_modified.replace('n/a', np.nan, regex=True)
big_chungus.head()
```

In a separate workbook we've merged budget information with the imdb tconst variable, so let's add that info to BIG CHUNGUS.

```python
budget_chungus = pd.read_csv('./etl_csvs/budget_chungus.csv')
```
