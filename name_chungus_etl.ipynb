```python
import pandas as pd
from glob import glob
import os
import numpy as np
```

```python
csv_files = glob("./people-data/*.csv")
csv_files
```

```python
d = {}
for file in csv_files:
    d[file] = pd.read_csv(file)
```

```python
csv_files_dict = {}
for filename in csv_files:
    filename_cleaned = os.path.basename(filename).replace(".csv", "").replace(".","_") #cleaning the file names
    filename_df = pd.read_csv(filename, index_col = 0)
    csv_files_dict[filename_cleaned] = filename_df
```

```python
imdb_name_basics_df = csv_files_dict['imdb_name_basics']
imdb_title_principals_df = csv_files_dict['imdb_title_principals']
```

```python
# Create new column of index in titles dataframe and setting nconst as new index. 
imdb_title_principals_df['tconst'] = imdb_title_principals_df.index
imdb_title_principals_df = imdb_title_principals_df.set_index('nconst')
imdb_title_principals_df.head()
```

```python
# Join both CSV files: 
name_frame = imdb_title_principals_df.join(imdb_name_basics_df)
name_frame.info()
```

```python
name_frame.head()
```

```python
name_frame['known_for_titles'] = name_frame['known_for_titles'].fillna('n/a')
name_frame['known_for_titles'] = name_frame['known_for_titles'].apply(lambda x: x.split(",") if x else x)
name_frame.head()
```

```python
f = lambda x: 'known_for_{}'.format(x + 1)

people_chungus = pd.DataFrame(
    name_frame['known_for_titles'].values.tolist(),
    name_frame.index, dtype=object
).fillna('n/a').rename(columns=f)
```

```python
kf1 = people_chungus['known_for_1']
kf2 = people_chungus['known_for_2']
kf3 = people_chungus['known_for_3']
kf4 = people_chungus['known_for_4']
people_chungus = pd.concat([kf1,kf2,kf3,kf4], axis = 1)
people_chungus.head()
```

```python
#Join the separated dataframe to original name_frame
name_frame = name_frame.join(people_chungus)
```

```python
#Drop original known_for columns
name_frame = name_frame.drop(columns= 'known_for_titles')
name_frame.head()
```

```python
#Replace 'n/a' string with Nan values:
name_chungus = name_frame.replace('n/a',np.nan, regex=True)
name_chungus.head()
```

And we have what we want, let's export as CSV:

```python
name_chungus.to_csv('imdb_name_chungus.csv')
```
